```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('voice.csv')
data.head()
```

```python
print('shape of data:', data.shape)
print('Total number of labels: {}'.format(data.shape[0]))
print('Total number of men: {}'.format(data[data.label=='male'].shape[0]))
print('Total number of Females: {}'.format(data[data.label=='female'].shape[0]))
```

```python
x = data.iloc[:,:-1]
print(data.shape)
print(x.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y = data.iloc[:,-1]
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix 
```

```python
svc_model = SVC()
svc_model.fit(x_train,y_train)
y_pred = svc_model.predict(x_test)
```

```python
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
p_grid = {'c':[0.1,1,10,100],'d':[1,0.1,0.01,0.001]}
```

```python
grid = GridSearchCV(SVC(),p_grid,refit=True,verbose=2)
grid.fit(x_train,y_train)
```
